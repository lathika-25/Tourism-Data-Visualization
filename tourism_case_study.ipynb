```python
import pandas as pd
import os 
df = pd.read_csv(r'C:\Users\lathi\OneDrive\Desktop\Tourism_CaseStudy\tourism_case_study.csv')

print(df.head())
```

```python
df_long = df.melt(id_vars=['Country'], var_name='Year', value_name='Arrivals')

# Convert Year to integer
df_long['Year'] = df_long['Year'].astype(int)

# Sort by Country and Year
df_long = df_long.sort_values(['Country', 'Year'])

# Show first 10 rows to check
print(df_long.head(10))
```

```python
df_long['Percent_Growth'] = df_long.groupby('Country')['Arrivals'].pct_change() * 100

# Fill NaN for the first year with 0
df_long['Percent_Growth'] = df_long['Percent_Growth'].fillna(0)

# Show first 10 rows to check
print(df_long.head(10))
```

```python
import plotly.express as px

#Choropleth map for Percent Growth
fig = px.choropleth(
    df_long,
    locations="Country",        # Column with country names
    locationmode="country names",  # Can also use "ISO-3" for future-proof
    color="Percent_Growth",     # Column for color intensity
    hover_name="Country",       # Show country on hover
    hover_data={"Year": True, "Arrivals": True, "Percent_Growth": True},
    animation_frame="Year",     # Shows animation year by year
    color_continuous_scale="Viridis",
    title="Yearly % Growth of Tourist Arrivals by Country"
)

fig.write_html("Tourism_Choropleth_Map.html")
print("Map saved! Open 'Tourism_Choropleth_Map.html' in your browser to see it.")
```
